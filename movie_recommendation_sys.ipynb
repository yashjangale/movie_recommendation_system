```python
import pandas as pd
```

```python
df=pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
# Checking for missing values in the entire dataset
print(df.isnull().sum())
```

```python
print(df[df.isnull().any(axis=1)])
```

```python
# Filling important columns' NaN values with an empty string
df['genres'] = df['genres'].fillna('')
df['keywords'] = df['keywords'].fillna('')
df['cast'] = df['cast'].fillna('')
df['director'] = df['director'].fillna('')
df['overview'] = df['overview'].fillna('')
df['tagline'] = df['tagline'].fillna('')
```

```python
print(df.isnull().sum())
```

```python
# Combine important features into a single column
df['combined_features'] = df['genres'] + ' ' + df['keywords'] + ' ' + df['cast'] + ' ' + df['director']
```

```python
# Display the combined features for validation
df['combined_features'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer object
cv = CountVectorizer(stop_words='english')
```

```python
# Convert combined features to a matrix of token counts
count_matrix = cv.fit_transform(df['combined_features'])
from sklearn.metrics.pairwise import cosine_similarity
# Compute the cosine similarity matrix
cosine_sim = cosine_similarity(count_matrix)
```

```python
# Display the similarity matrix shape
print(cosine_sim.shape)
```

```python
# Function to get the index of a movie from its title
def get_index_from_title(title):
    return df[df['title'] == title].index[0]

# Function to get the title of a movie from its index
def get_title_from_index(index):
    return df.iloc[index]['title']
```

```python
# Function to recommend movies
def recommend_movies(movie_title, num_recommendations=10):
    movie_index = get_index_from_title(movie_title)
    similarity_scores = list(enumerate(cosine_sim[movie_index]))
    sorted_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    # Get top recommendations
    recommended_movie_indices = [x[0] for x in sorted_scores[1:num_recommendations+1]]
    return [get_title_from_index(idx) for idx in recommended_movie_indices]
```

```python
# Testing
movie_name = "The Dark Knight"  
recommended_movies = recommend_movies(movie_name)

print(f"Movies similar to '{movie_name}':")
for i, movie in enumerate(recommended_movies, 1):
    print(f"{i}. {movie}")
```
